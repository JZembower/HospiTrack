# Geocoding Data:

Geocoding the provider datasets:

```python
# Merge the inpatient and outpatient datasets
merged_provider = pd.merge(
    medicare_inpatient_provider,
    medicare_outpatient_provider,
    on=['provider_ccn', 'provider_org_name', 'provider_city', 'provider_street_address', 'provider_state_abbreviation', 'provider_state_fips', 'provider_zip5', 'provider_ruca_code', 'provider_ruca_description'],
    how='outer',
    suffixes=('_inpatient', '_outpatient')
)
merged_provider.head()
```

```python
# Create full address for further query in geospatial data
merged_provider['full_address'] = (
    merged_provider['provider_street_address'].str.replace(r'\b(Box|Suite)\s+\d+\b', '', regex=True).str.strip() +
    ', ' +
    merged_provider['provider_city'] +
    ', ' +
    merged_provider['provider_state_abbreviation'] +
    ', ' +
    merged_provider['provider_zip5'].astype(str)
)

# Output cleaned dataset for extrapolating
merged_provider[['full_address']].head()
```

```python
merged_provider['provider_state_abbreviation']
```

```python
illinois_data = merged_provider[merged_provider['provider_state_abbreviation'] == 'IL']
```

```python
geolocator = Nominatim(user_agent="er_wait_times_project")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

geocode_cache = {}

def get_lat_lon(addr):
    if addr in geocode_cache:
        return geocode_cache[addr]
    try:
        location = geocode(addr)
        if location:
            geocode_cache[addr] = (location.latitude, location.longitude)
            return geocode_cache[addr]
    except Exception as e:
        print(f"Geocoding error for {addr}: {e}")
    return (None, None)

illinois_data[['lat', 'lon']] = illinois_data['full_address'].apply(lambda addr: pd.Series(get_lat_lon(addr)))

illinois_data[['provider_org_name', 'full_address', 'lat', 'lon']].head()
```
